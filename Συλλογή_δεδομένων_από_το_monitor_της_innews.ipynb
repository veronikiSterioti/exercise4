## ΒΗΜΑ 2. Ορισμός "θεματικών" κατηγοριών

Οι κατηγορίες (Categories) πρακτικά είναι οι όροι της αναζήτησης που θέτουμε ώστε να φιλτράρουμε τα δεδομένα της innews για να πάρουμε μόνο αυτά που μας ενδιαφέρουν.

Αρχικά, στις Κατηγορίες βρίσκονται οι προκαθορισμένες "θεματικές" κατηγορίες που έχει φτιάξει η innews για εμας.

Δεν μπορούμε να δημιουργήσουμε νέες κατηγορίες, αλλά μπορούμε να τροποποιήσουμε τις υπάρχουσες.

### Τροποποίηση των Κατηγοριών, σύμφωνα με το θέμα που θέλουμε να μελετήσουμε στην εργασία μας

- Από το αριστερό Menu πάμε στα **Settings** -> **Categories**
- Διαλέγουμε μια Κατηγορία που δεν θέλουμε και πατάμε δεξια το **Edit**
- Αλλάζουμε τον Τίτλο ώστε να αντιπροσωπεύει την κατηγορία που θέλουμε
- Θέτουμε τα keywords βάσει των οποίων θα φιλτράρονται τα δεδομένα της innews ώστε να είναι σχετικά με το θέμα της Κατηγορίας μας
- Πατάμε Save

**Σημείωση:** Οι κανόνες για τον ορισμό των keywords περιγράφονται στο "How it works" που βρίσκεται δίπλα στον τίτλο CATEGORIES (στην κορυφή της σελίδας Settings -> Categories, όπου εμφανίζονται όλες οι κατηγορίες)

## ΒΗΜΑ 3. Επιλογή δεδομένων

Από το αριστερό Menu πάμε στο **Categories** και επιλέγουμε την κατηγορία που θέλουμε.

Στη σελίδα εμφανίζονται όλα τα αποτελέσματα που περιλαμβάνουν τα keywords που έχουμε θέσει για τη συγκεκριμένη κατηγορία.

Τα αποτελέσματα αφορούν τη **σημερινή ημέρα** και προέρχονται από **όλα τα μέσα** που συλλέγει η Innews.

### Επιλογή χρονικού διαστήματος

Στην κορυφή της σελίδας μπορούμε να επιλέξουμε το χρονικό διάστημα που μας ενδιαφέρει:
- Day (αποτελέσματα από τη σημερινή ημέρα)
- 7days (αποτελέσματα από τις τελευταίες 7 ημέρες)
- 15days
- 30days
- Choose date (ορίζουμε εμείς το χρονικό διάστημα που μας ενδιαφέρει)


### Επιλογή (είδους) μέσων

Ακριβώς κάτω από τις επιλογές του χρονικού διαστήματος, δεξιά υπάρχει το πράσινο κουμπί με τίτλο **Edit filters**.

Από το menu **Media Types** επιλέγουμε το **Web** και πατάμε (στο κάτω μέρος της σελίδας) **Save filters**.

H σελίδα των αποτελεσμάτων ενημερώνεται ώστε να ανταποκρίνεται στις επιλογές που κάναμε.
